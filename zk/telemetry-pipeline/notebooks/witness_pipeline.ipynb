## Witness generation\nLoad production-style accumulator output and transform to Circom input.


```python
import json, hashlib
with open('../data/sample_telemetry.json') as f:
    t=json.load(f)
values=t['lotA']+t['lotB']
sha=hashlib.sha512(bytes(values)).digest()
bits=[(b>>i)&1 for b in sha for i in range(7,-1,-1)]
len(bits), sum(bits)
```

```python
from graphviz import Digraph
g = Digraph('telemetry_pipeline')
g.attr(rankdir='LR')
g.node('T', 'telemetry\n(2x16 samples)')
g.node('A', 'accumulator\nSVE2 fold')
g.node('W', '512-bit witness')
g.node('S', 'SNARK proof')
g.node('C', 'on-chain syslog')
g.edges(['TA','AW','WS','SC'])
g
```
